```python
Age_stats = df['Age'].describe()
print(Age_stats)
```

```python
variable = 'Age'

# Plot the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df[variable])
plt.title(f'Boxplot for {variable}')
plt.xlabel(variable)
plt.show()
```

```python

heart_rate_stats = df['HeartRate'].describe()
print(heart_rate_stats)
```

```python
# Plot boxplot of heart rate grouped by another variable, e.g., 'activity'
sns.boxplot(x='Cholesterol', y='HeartRate', data=df)

# Formatting the plot
plt.title('Heart Rate by Age')
plt.xlabel('Age')
plt.ylabel('Heart Rate (bpm)')
plt.show()
```

```python
print(df['Cholesterol'].unique())  # Check unique values
print(df['Cholesterol'].value_counts())  # Check frequency of each value
```

```python
# Define categories for Cholesterol levels
bins = [0, 200, 240, 500]  # Define your custom bin edges
labels = ['Low', 'Medium', 'High']  # Define labels for the bins
df['Cholesterol_Category'] = pd.cut(df['Cholesterol'], bins=bins, labels=labels, right=False)

# Now, plot the boxplot with the categorized Cholesterol values
sns.boxplot(x='Cholesterol_Category', y='HeartRate', data=df)
plt.title('Heart Rate by Cholesterol Category')
plt.xlabel('Cholesterol Category')
plt.ylabel('Heart Rate (bpm)')
plt.show()
```

```python
# Remove rows with missing or invalid values (if any)
df = df.dropna(subset=['Cholesterol', 'HeartRate'])

# Optionally, filter out any rows with invalid cholesterol values
df = df[df['Cholesterol'] > 0]  # Remove negative or zero cholesterol values if necessary
```

```python
sns.scatterplot(x='Cholesterol', y='HeartRate', data=df)
plt.title('Heart Rate vs. Cholesterol')
plt.xlabel('Cholesterol')
plt.ylabel('Heart Rate (bpm)')
plt.show()
```
